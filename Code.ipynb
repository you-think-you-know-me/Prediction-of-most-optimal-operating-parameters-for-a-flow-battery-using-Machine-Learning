```python
#For Area=400cm^2 and Current=60J
import pandas as pd
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

# Load the training and testing data from Excel sheets
train_data = pd.read_excel('Train.xlsx')
test_data = pd.read_excel('Test.xlsx')

# Separate the input features (SF, Capacity) and output (V) for training and testing
X_train = train_data[['SF', 'Capacity']]
y_train = train_data['V']
X_test = test_data[['SF', 'Capacity']]
y_test = test_data['V']

# Dictionary of models to test
models = {
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42),
    'Linear Regression': LinearRegression(),
    'Gradient Boosting': GradientBoostingRegressor(random_state=42),
    'Support Vector Machine': SVR(),
    'Neural Network': MLPRegressor(random_state=42, max_iter=500)
}

# Loop through each model, train, test, evaluate, and plot
for name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)

    # Predict on the test data
    y_pred = model.predict(X_test)

    # Evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    # Print evaluation metrics
    print(f'{name} Results:')
    print(f'Mean Squared Error: {mse}')
    print(f'R-squared: {r2}')
    print('-' * 50)

    # Plotting the actual vs predicted values in separate plots
    plt.figure(figsize=(10, 6))
    sns.lineplot(x=test_data['Capacity'], y=y_test, label='Actual V', marker='o', color='blue')
    sns.lineplot(x=test_data['Capacity'], y=y_pred, label='Predicted V', marker='o', color='red')
    plt.xlabel('Capacity')
    plt.ylabel('V')
    plt.title(f'Actual vs Predicted V - {name}')
    plt.legend()
    plt.show()
```
